# [dbo].[sp_read]

```sql
/****** Object:  StoredProcedure [dbo].[sp_read]    Script Date: 21/10/2023 10:59:49 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_read]
    @Id INT
AS
BEGIN
    SELECT
        Id_Usuario,
        Nombre,
        Apellido,
        Valor_Vivienda,
        Tasa,
        Plazo,
        PorcentajeCuotaInicial,
        ValorCuotaInicial,
        Credito,
        ValorBeneficio,
        -- Calcular la cuota sin beneficio
        (Valor_Vivienda * 0.0115 * POWER(1 + 0.0115, Plazo)) / (POWER(1 + 0.0115, Plazo) - 1) AS CuotaSinBeneficio,
        -- Calcular la cuota con beneficio
        ((Valor_Vivienda * 0.0115 * POWER(1 + 0.0115, Plazo)) / (POWER(1 + 0.0115, Plazo) - 1)) - ValorBeneficio AS CuotaConBeneficio
    FROM
        Usuarios
    WHERE
        Id_Usuario = @Id;
END
GO
```

# [dbo].[sp_update]

```sql
/****** Object:  StoredProcedure [dbo].[sp_update]    Script Date: 21/10/2023 10:59:49 a. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_update]
    @Id INT,
    @Nombre VARCHAR(50),
    @Apellido VARCHAR(50),
    @Valor_Vivienda DECIMAL(18, 2),
    @Tasa DECIMAL(18, 2),
    @Plazo INT,
    @PorcentajeCuotaInicial DECIMAL(18, 2),
    @ValorCuotaInicial DECIMAL(18, 2),
    @Credito DECIMAL(18, 2),
    @ValorBeneficio DECIMAL(18, 2),
    @CuotaSinBeneficio DECIMAL(18, 2),
    @CuotaConBeneficio DECIMAL(18, 2)
AS
BEGIN
    UPDATE Usuarios
    SET
        Nombre = @Nombre,
        Apellido = @Apellido,
        Valor_Vivienda = @Valor_Vivienda,
        Tasa = @Tasa,
        Plazo = @Plazo,
        PorcentajeCuotaInicial = @PorcentajeCuotaInicial,
        ValorCuotaInicial = @ValorCuotaInicial,
        Credito = @Credito,
        ValorBeneficio = @ValorBeneficio,
        CuotaSinBeneficio = @CuotaSinBeneficio,
        CuotaConBeneficio = @CuotaConBeneficio
    WHERE Id_Usuario = @Id;
END
GO
```
